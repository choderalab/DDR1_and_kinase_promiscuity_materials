```python
import matplotlib.pyplot as plt
import pandas as pd

import matplotlib
import numpy as np

%matplotlib inline
```

```python
Drewry_data = pd.read_csv('Drewry-data.csv')
```

```python
Drewry_data
```

```python
count_1 = Drewry_data.apply(lambda x: (x > 1).sum(),axis=0)
count_10 = Drewry_data.apply(lambda x: (x > 10).sum(),axis=0)
count_25 = Drewry_data.apply(lambda x: (x > 25).sum(),axis=0)
count_50 = Drewry_data.apply(lambda x: (x > 50).sum(),axis=0)
count_75 = Drewry_data.apply(lambda x: (x > 75).sum(),axis=0)
count_90 = Drewry_data.apply(lambda x: (x > 90).sum(),axis=0)
```

```python
count_data = {'> 1% Inh':count_1,
       '> 10% Inh':count_10,
       '> 25% Inh':count_25,
       '> 50% Inh':count_50,
       '> 75% Inh':count_75,
       '> 90% Inh':count_90}
```

```python
count_df = pd.DataFrame(data=count_data)
```

```python
df_75 = pd.DataFrame(count_df['> 75% Inh'])
df_75.transpose().to_csv('data-Drewry-75inh.csv')

df_90 = pd.DataFrame(count_df['> 90% Inh'])
df_90.transpose().to_csv('data-Drewry-90inh.csv')
```

```python
import seaborn as sns
sns.set_context("poster")
sns.set_style("ticks")
```

```python
colors = matplotlib.cm.jet_r(np.linspace(0.15, 0.6, 175))
```

```python
plt.figure(figsize=(6,3))

plt.axvline(x=count_df['> 90% Inh']['DDR2'],color='0.7',alpha=0.7,label='DDR2',zorder=0)
plt.axvline(x=count_df['> 90% Inh']['PDGFRA'],color='0.7',alpha=0.7,label='PDGFRA',zorder=0)
plt.axvline(x=count_df['> 90% Inh']['CSF1R'],color='0.7',alpha=0.7,label='CSF1R',zorder=0)
plt.axvline(x=count_df['> 90% Inh']['YSK4'],color='0.7',alpha=0.7,label='YSK4',zorder=0)
plt.axvline(x=count_df['> 90% Inh']['DDR1'],color='0.7',alpha=0.7,linestyle='--',label='DDR1',zorder=0)
plt.axvline(x=count_df['> 90% Inh']['MEK5'],color='0.7',alpha=0.7,label='MEK5',zorder=0)
plt.axvline(x=count_df['> 90% Inh']['KIT'],color='0.7',alpha=0.7,label='KIT',zorder=0)
plt.axvline(x=count_df['> 90% Inh']['PDGFRB'],color='0.7',alpha=0.7,label='PDGFRB',zorder=0)

axes = count_df['> 90% Inh'].plot(kind='hist',color='white', bins=150)

k=0
#for rect in axes.patches:
if rect.get_height() > 0:
    for rect in axes.patches:
        rect.set_color(colors[k])
        k+=1

plt.xlabel('Number of ligands')
plt.ylabel('Number of kinases')
plt.legend(bbox_to_anchor=(1.01, 1.01))

plt.savefig('promiscuity_plot_rainbow_Inh90.eps',format='eps',dpi=300,bbox_inches='tight')
```

```python
Davis_data = pd.read_csv('Davis_2011_transpose.csv')
```

```python
Davis_data = Davis_data.fillna(10000)
```

```python
Davis_data
```

```python
Davis_data.values.max()
```

```python
#Here intstead of counting by percent affinity assuming nM units
Dcount_5000 = Davis_data.apply(lambda x: (x < 5000).sum(),axis=0) #lowest affinity
Dcount_1000 = Davis_data.apply(lambda x: (x < 1000).sum(),axis=0)
Dcount_500 = Davis_data.apply(lambda x: (x < 500).sum(),axis=0)
Dcount_100 = Davis_data.apply(lambda x: (x < 100).sum(),axis=0)
Dcount_50 = Davis_data.apply(lambda x: (x < 50).sum(),axis=0) #highest affinity
```

```python
Dcount_data = {'< 5000 nM': Dcount_5000,
       '< 1000 nM': Dcount_1000,
       '< 500 nM': Dcount_500,
       '< 100 nM': Dcount_100,
       '< 50 nM': Dcount_50}
```

```python
Dcount_df = pd.DataFrame(data=Dcount_data)
```

```python
colors = matplotlib.cm.jet_r(np.linspace(0.15, 0.6, 30))
```

```python
plt.figure(figsize=(6,3))

plt.axvline(x=Dcount_df['< 50 nM']['DDR2'],color='0.7',alpha=0.7,label='DDR2',zorder=0)
plt.axvline(x=Dcount_df['< 50 nM']['MEK5'],color='0.7',alpha=0.7,label='MEK5',zorder=0)
plt.axvline(x=Dcount_df['< 50 nM']['YSK4'],color='0.7',alpha=0.7,label='YSK4',zorder=0)
plt.axvline(x=Dcount_df['< 50 nM']['DDR1'],color='0.7',alpha=0.7,linestyle='--',label='DDR1',zorder=0)
plt.axvline(x=Dcount_df['< 50 nM']['PDGFRA'],color='0.7',alpha=0.7,label='PDGFRA',zorder=0)
plt.axvline(x=Dcount_df['< 50 nM']['CSF1R'],color='0.7',alpha=0.7,label='CSF1R',zorder=0)
plt.axvline(x=Dcount_df['< 50 nM']['KIT'],color='0.7',alpha=0.7,label='KIT',zorder=0)
plt.axvline(x=Dcount_df['< 50 nM']['PDGFRB'],color='0.7',alpha=0.7,label='PDGFRB',zorder=0)

axes = Dcount_df['< 50 nM'].plot(kind='hist',color='white', bins=30)

k=0
#for rect in axes.patches:
if rect.get_height() > 0:
    for rect in axes.patches:
        rect.set_color(colors[k])
        k+=1

plt.title('Davis 2011')
plt.xlabel('Number of ligands');
plt.ylabel('Number of kinases');
plt.legend(bbox_to_anchor=(1.01, 1.01));


plt.savefig('promiscuity_plot_Davis.png',dpi=300,bbox_inches='tight')
```
